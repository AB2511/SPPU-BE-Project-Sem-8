## Problem Statement:

Study morphology concepts and implement Add-Delete table to analyze word formation from morphemes.

## Morphology

Morphology is the study of how words are formed from smaller meaningful units called morphemes.

Example:
- running → run + ing
- cats → cat + s

Add-Delete Table is used to generate different word forms by:
- Adding suffixes
- Deleting suffixes

```python
words = ["run", "play", "walk"]

print("Base Words:", words)
```

```python
suffixes = ["s", "ing", "ed"]

def add_suffix(word):
    forms = []
    for suffix in suffixes:
        forms.append(word + suffix)
    return forms

for word in words:
    print(f"{word} → {add_suffix(word)}")
```

```python
def remove_suffix(word):
    if word.endswith("ing"):
        return word[:-3]
    elif word.endswith("ed"):
        return word[:-2]
    elif word.endswith("s"):
        return word[:-1]
    return word

test_words = ["running", "played", "cats"]

for word in test_words:
    print(f"{word} → {remove_suffix(word)}")
```

```python
def add_delete_table(word):
    forms = {}
    
    # Add forms
    forms['added'] = [word + suffix for suffix in suffixes]
    
    # Delete forms
    forms['deleted'] = remove_suffix(word)
    
    return forms

for word in ["run", "play", "cat"]:
    print(f"\nWord: {word}")
    print(add_delete_table(word))
```

```python
sentence = "running cats played"

words = sentence.split()

print("Original Sentence:", sentence)

processed = [remove_suffix(word) for word in words]

print("After Morphological Processing:", " ".join(processed))
```
