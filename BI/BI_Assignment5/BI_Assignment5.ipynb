## Title: Perform Data Classification using Classification Algorithm
## Aim: To perform data classification using Naive Bayes algorithm.

```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Step 2: Load dataset
data = load_iris()
X = data.data
Y = data.target
```

```python
# Step 3: Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Step 4: Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Step 5: Train model
model = GaussianNB()
model.fit(X_train, Y_train)
```

```python
# Step 6: Prediction
predictions = model.predict(X_test)
```

```python
# Step 7: Confusion Matrix
cm = confusion_matrix(Y_test, predictions)
print("Confusion Matrix:\n", cm)
```

```python
# Step 8: Classification Report
print("\nClassification Report:\n", classification_report(Y_test, predictions))
```

```python
sns.heatmap(cm, annot=True, cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```

### Result: 
- The Naive Bayes model was successfully trained and tested.
- The model achieved 100% accuracy on the test dataset.

### Conclusion: Naive Bayes classification algorithm effectively classifies data with high accuracy and can be used for predictive analysis.
