### Title:

Implementation of Bag of Words, TF-IDF and Word2Vec Embeddings

### Problem Statement:

Perform Bag-of-Words approach (count occurrence, normalized count occurrence), TF-IDF on dataset and create word embeddings using Word2Vec.

```python
# !pip install pandas scikit-learn gensim

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
```

```python
df = pd.read_csv("data.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
# Combine relevant columns into a single text column
df['text'] = (
    df['Make'].fillna('') + " " +
    df['Model'].fillna('') + " " +
    df['Engine Fuel Type'].fillna('') + " " +
    df['Transmission Type'].fillna('') + " " +
    df['Driven_Wheels'].fillna('')
)

# Ensure all values are strings
df['text'] = df['text'].astype(str)

df['text'].head()
```

## PART 1: BAG OF WORDS

```python
count_vectorizer = CountVectorizer()

X_count = count_vectorizer.fit_transform(df['text'])

print("BoW Matrix Shape:", X_count.shape)
print("Sample Features:", count_vectorizer.get_feature_names_out()[:10])
```

```python
import pandas as pd

bow_df = pd.DataFrame(
    X_count.toarray()[:5],
    columns=count_vectorizer.get_feature_names_out()
)

bow_df.head()
```

## PART 2: NORMALIZED COUNT (TF)

```python
tf_vectorizer = TfidfVectorizer(use_idf=False, norm='l2')

X_tf = tf_vectorizer.fit_transform(df['text'])

print("TF Matrix Shape:", X_tf.shape)
```

## PART 3: TF-IDF

```python
tfidf_vectorizer = TfidfVectorizer()

X_tfidf = tfidf_vectorizer.fit_transform(df['text'])

print("TF-IDF Matrix Shape:", X_tfidf.shape)
```

```python
print("Top Features:", tfidf_vectorizer.get_feature_names_out()[:10])
```

## PART 4: WORD2VEC

```python
sentences = [text.lower().split() for text in df['text']]

print("Sample Tokenized Sentence:")
print(sentences[0])
```

```python
model = Word2Vec(
    sentences=sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=2
)
```

```python
print("Vector for 'toyota':\n")

if 'toyota' in model.wv:
    print(model.wv['toyota'])
else:
    print("Word not found in vocabulary")
```

```python
if 'toyota' in model.wv:
    print("Words similar to 'toyota':\n")
    print(model.wv.most_similar('toyota'))
```

## Conclusion

- Bag of Words counts word frequency in documents.
- TF normalizes the frequency values.
- TF-IDF gives importance to rare words.
- Word2Vec generates vector representations capturing semantic meaning.

The structured dataset was converted into text format before applying NLP techniques.
